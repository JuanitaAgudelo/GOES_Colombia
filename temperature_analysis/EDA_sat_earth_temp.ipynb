```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2019

```python
sat_2 = pd.read_csv('t_sat_2019_1grad.csv', index_col=[0])
ear2 =  pd.read_csv('estacion_data_temperatura_206__20190101_20190131.csv')
```

```python
sat_2['fecha'] = pd.to_datetime(sat_2['fecha'], format='%Y-%m-%d %H:%M:%S')
sat_2_1 = sat_2[sat_2['fecha'].dt.month == 1]
sat_2_1['temperatura'] = sat_2_1.temperatura.apply(lambda x: x - 273.15)
```

```python
print('Datos faltantes enero 2019 ', (sat_2_1['temperatura'].isna().sum()/len(sat_2_1)) * 100)
```

```python
ear2['fecha'] = pd.to_datetime(ear2['fecha_hora'], format='%Y-%m-%d %H:%M:%S')
ear2['Temperatura'] = ear2['Temperatura'].replace(-999.0, np.nan)

temp2 = []
n=0
for i in range(744):
    temp2.append(ear2[n:n+6]['Temperatura'].mean())
    n+=60
```

```python
temperatura2 = pd.DataFrame(columns = ['Fecha', 'LST', 'Temperatura'])
temperatura2['Fecha'] = sat_2_1['fecha']
temperatura2['LST'] = sat_2_1['temperatura']
temperatura2['Temperatura'] = temp2
```

```python
fig, ax = plt.subplots(figsize = (10,4))
plt.plot(temperatura2['Fecha'], temperatura2['Temperatura'], '.', color = 'blue', label='Temperatura SIATA')
plt.plot(temperatura2['Fecha'], temperatura2['LST'], '.', color='orange', label='LST')
plt.title('LST y temperatura SIATA Medellín enero 2019')
plt.ylabel('Temperatura [°C]')
plt.xlabel('Fecha')
plt.legend()
plt.show();
```

```python
week1 = temperatura2[(temperatura2['Fecha'].dt.day >= 1) & (temperatura2['Fecha'].dt.day <= 7)]
print('datos faltantes ', week1['LST'].isna().sum()/len(week1) * 100, '%')
fig, ax = plt.subplots(figsize=(10,5))
plt.plot(week1['Fecha'], week1['Temperatura'], '-.', color = 'blue', label=('SIATA'))
plt.plot(week1['Fecha'], week1['LST'], '-.', color='orange', label = ('LST'))
plt.title('Temperatura SIATA y LST primer semana de enero 2019 por horas')
plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.legend()
plt.grid()
plt.show();
```

## 2020 

```python
sat_1 = pd.read_csv('t_sat_01.csv', index_col=[0])
ear = pd.read_csv('estacion_data_temperatura_229__20200101_20200131.csv')
```
